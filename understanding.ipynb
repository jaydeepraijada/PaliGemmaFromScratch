- Architecture
    1. Image Encoder (SigLIP - "shape optimized" VitSo400m Img Enc)
        - Embedding
            - Conv2d to extract patches
            - convert patches to embeddings
            - create and add positional embeddings
        - Encoder block 
            - Skip connection (added post self attention)
            - Layernorm
            - Self Attention
            - Skip Connection (added post MLP)
            - Layernorm
            - MLP
        - Post Layer Norm

    2. Decoder Only Language Model Gemma-2B

    3. Linear Layer projecting SigLIP's output tokens into the same dimensions as Gemma 2B's tokens, so they can be concatenated
